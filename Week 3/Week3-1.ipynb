```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
@tf.function
def predict(X, W, b):
    return tf.nn.sigmoid(tf.matmul(X, W) + b)
```

```python
@tf.function
def L(y, y_hat):
    return tf.reduce_mean((y - y_hat)**2)
```

```python
X = tf.constant([[0.0, 0], [0, 1], [1, 0], [1, 1]])
y = tf.constant([[0.0], [0], [0], [1]])
W = tf.Variable(tf.random.normal(shape=(2, 1), mean=0.0, stddev=1.0)) 
b = tf.Variable(tf.random.normal(shape=(1,), mean=0.0, stddev=0.1)) 
alpha = 0.1
for it in range(500):
    with tf.GradientTape() as t:
        current_loss = L(y, predict(X, W, b))
    print("it", it, current_loss)
    dW, db = t.gradient(current_loss, [W, b])
    W.assign_sub(alpha * dW)
    b.assign_sub(alpha * db)
```

```python
y_hat = predict(X, W, b)
predicted_labels = (y_hat > 0.5).numpy().astype(int)

df_results = pd.DataFrame({
    'Predicted Probability': np.array(y_hat.numpy()).flatten(),
    'Predicted Label': np.array(predicted_labels).flatten()
})

print(df_results)
l = L(y, y_hat)
print(l)
```
