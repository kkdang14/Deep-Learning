```python
import tensorflow as tf
import numpy as np
```

```python
@tf.function
def layer1(X, W, B1):
    return tf.nn.relu(tf.matmul(X, W) + B1)

@tf.function
def layer2(X, W, B2):
    return tf.nn.sigmoid(tf.matmul(X, W) + B2)

@tf.function
def predict(X, W1, B1, W2, B2):
    return layer2(layer1(X, W1, B1), W2, B2)

@tf.function
def L(y, y_hat):
    loss = y * tf.math.log(y_hat)
    return -tf.reduce_mean(tf.reduce_sum(loss, axis=1))  #ma tran cong tung dong lai 

X = tf.constant([[0.0, 0], [0, 1], [1, 0], [1, 1]])
y = tf.constant([[0.0,], [1], [1], [0]])
W1 = tf.Variable(tf.random.normal((2, 2)))
B1 = tf.Variable([0.0, 0.0])
W2 = tf.Variable(tf.random.normal((2, 1)))    
B2 = tf.Variable([0.0])
alpha = 0.1
for it in range(500):
    with tf.GradientTape() as t:
        current_loss = L(y, predict(X, W1, B1, W2, B2))
    print("it", it, current_loss)
    dW1, dB1, dW2, dB2 = t.gradient(current_loss, [W1, B1, W2, B2])
    W1.assign_sub(alpha * dW1)
    B1.assign_sub(alpha * dB1)
    W2.assign_sub(alpha * dW2)
    B2.assign_sub(alpha * dB2)
    if it % 10 == 0:
        print(f'Epoch {it}, Loss: {current_loss.numpy()}')
```

```python
y_hat = predict(X, W1, B1, W2, B2)
y_pred_labels = tf.argmax(y_hat, axis=1).numpy()
y_true_labels = tf.argmax(y, axis=1).numpy()
print(y_pred_labels)
print(y_true_labels)
    
accuracy = np.mean(y_pred_labels == y_true_labels)
print(accuracy)
```
